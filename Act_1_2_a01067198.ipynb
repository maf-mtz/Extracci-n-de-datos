### Filtro 1: CVE_CLPV valores 1000 a 2000

```python
#df['CVE_CLPV'].astype(str).astype(int)
```

```python
#Filtro por comparacion
filtro1=df[(df["CVE_CLPV"] >=1000) & (df["CVE_CLPV"] <=2000)]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable1_2.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable1_2.csv")
```

### Filtro 2: CVE_VEND Todas las claves excepto "5" y "6"

```python
#Filtro not
filtro2=df[~(df["CVE_VEND"] ==5) & ~(df["FECHA_DOC"] == 6)]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Entregable2_2.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable2_2.csv")
```

### Filtro 3: FECHA_ENT Fechas de '28/02/2022'

### Filtro 4: CAN_TOT Cantidades menores a '5951.7' o Status 'E'
