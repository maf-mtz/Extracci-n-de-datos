## **FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro8=df[df["COSTO_TOTAL_CALCULADO"] > 1500]
filtro8

#filtrado por aquellas filas en donde el 'COSTO_TOTAL_CALCULADO' sea mayor a 1500
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Entregable8.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable8.csv")
```

## **FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro9=df[(df["COSTO_TOTAL_CALCULADO"] < 5000) & (df["FECHA_DOC"] == "2022-12-23")]
filtro9

#Aquellas ventas en donde el costo total es menor a 5 mil y ademas, se realizaron el 23 de diciembre de 2022.
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable9.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable9.csv")
```

```python
#Filtro o
filtro10=df[(df["COSTO_TOTAL_CALCULADO"] > 10000) |  (df["FECHA_DOC"] == "2022-12-23")]
filtro10

#Filtrado por aquellas ventas en donde el costo total fue mayor a 10 mil o que se realizaron el 23 de diciembre del 2022
```

```python
#Convertir archivo filtrado a CSV
filtro10.to_csv("Entregable10.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable10.csv")
```

```python
#Filtro not
filtro11=df[~(df["COSTO_TOTAL_CALCULADO"] <= 10000) & ~(df["FECHA_DOC"] == "2022-12-25")]
filtro11

#Las ventas en donde no hubo un costo total menor o igual a 10 mil y que no se realizaron el 25 de diciembre de 2022
```

```python
#Convertir archivo filtrado a CSV
filtro11.to_csv("Entregable11.csv")

#descardar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Entregable11.csv")
```
